```python
import pandas as pd

xlsx_file = 'Weather+Underground+-+La+Madeleine,+FR(1).xlsx'
excel = pd.ExcelFile(xlsx_file)

for sheet_name in excel.sheet_names:
    df = excel.parse(sheet_name)
    csv_filename = f"{sheet_name}_FR.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Feuille '{sheet_name}' enregistrée sous '{csv_filename}'")
```
