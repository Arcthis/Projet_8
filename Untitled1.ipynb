```python
import pandas as pd
import requests
import json

df1 = pd.read_excel("source1.xlsx")
df2 = pd.read_excel("source2.xlsx")

records1 = df1.to_dict(orient="records")
records2 = df2.to_dict(orient="records")

url = "https://your-aws-link.com/data.json"
response = requests.get(url)
json_data = response.json()

with open("output1.json", "w") as f:
    json.dump(records1, f, indent=2)

with open("output2.json", "w") as f:
    json.dump(records2, f, indent=2)

with open("output3.json", "w") as f:
    json.dump(json_data, f, indent=2)
```
